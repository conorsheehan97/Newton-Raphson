```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.sin(x) - 0.5


plt.plot(x, y, label=r'$\sin(x) - 0.5$')
plt.axhline(0, color='red', linestyle='--', label='y=0')
plt.axvline(0.524,color='red', linestyle='--', label= 'Newton Raphson Root ')
#Adding labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $\sin(x) - 0.5$')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

As we can see above the root we found, was one of many. We started with an initial guess of 1, and found the closest root $\pi / 6$. Cool to see that the root found depends so much on the initial guess.
